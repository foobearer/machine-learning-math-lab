# Gradient Descent in Linear Regression

---

## What is Gradient Descent?

In the Cost Function notebook, we learned how to **measure** how wrong our model is using $J(w, b)$.

Now the question is: **how do we automatically find the best `w` and `b` that make the cost as small as possible?**

The answer is **Gradient Descent**, one of the most important algorithms in all of machine learning.

---

## The Intuition: Rolling a Ball Down a Hill

Imagine you are standing somewhere on a hilly landscape and you are **blindfolded**. Your goal is to reach the **lowest valley**.

What do you do? You feel the ground beneath your feet and take a small step in whichever direction feels **downhill**.

You repeat this —> step, feel, step, feel, you do this until you can no longer go any lower. That is the **minimum**.

Gradient descent does **exactly this**, but mathematically:

| Step | What it means mathematically |
|------|------------------------------|
| Feel the slope of the ground | Compute the **partial derivative** (gradient) of the cost |
| Step downhill | Subtract a fraction of the gradient from `w` and `b` |
| How big a step? | Controlled by the **learning rate** $\alpha$ |
| Repeat until flat | Loop until the cost stops decreasing |

---

## Recap: The Model and Cost Function

Our prediction model:
$$f_{w,b}(x) = wx + b$$

Our cost function (Mean Squared Error):
$$J(w,b) = \frac{1}{2m} \sum_{i=0}^{m-1} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)^2$$

Our goal: find the values of $w$ and $b$ that **minimize** $J(w, b)$.

---

## The Gradient Descent Update Rules

$$\boxed{\begin{align*}
w &:= w - \alpha \frac{\partial J(w,b)}{\partial w} \\
b &:= b - \alpha \frac{\partial J(w,b)}{\partial b}
\end{align*}}$$

**Repeat** these two updates simultaneously until convergence.

| Symbol | Meaning |
|--------|---------|
| $w$, $b$ | The parameters we are trying to optimize |
| $\alpha$ | **Learning rate** — how big each step is (e.g. 0.01) |
| $\frac{\partial J}{\partial w}$ | Partial derivative of cost w.r.t. $w$ — the slope in the $w$ direction |
| $\frac{\partial J}{\partial b}$ | Partial derivative of cost w.r.t. $b$ — the slope in the $b$ direction |
| $:=$ | Assignment — we update and overwrite the old value |

> **Why subtract?** If the slope is positive (uphill), subtracting moves us in the downhill direction. If the slope is negative (downhill already), subtracting a negative means adding — again moving toward the minimum.

> **Simultaneously** means: compute both partial derivatives using the **current** values of $w$ and $b$ **before** updating either one.

---

## Deriving the Gradients: Step-by-Step Calculus

We need to find $\frac{\partial J}{\partial w}$ and $\frac{\partial J}{\partial b}$.

### Starting point
$$J(w,b) = \frac{1}{2m} \sum_{i=0}^{m-1} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)^2$$

Substituting $f_{w,b}(x^{(i)}) = wx^{(i)} + b$:
$$J(w,b) = \frac{1}{2m} \sum_{i=0}^{m-1} \left(wx^{(i)} + b - y^{(i)}\right)^2$$

---

### Deriving $\frac{\partial J}{\partial w}$

**Step 1:** Apply the chain rule. The $\frac{1}{2m}$ and $\sum$ pass through, and we differentiate the squared term:
$$\frac{\partial J}{\partial w} = \frac{1}{2m} \sum_{i=0}^{m-1} 2\left(wx^{(i)} + b - y^{(i)}\right) \cdot \frac{\partial}{\partial w}\left(wx^{(i)} + b - y^{(i)}\right)$$

**Step 2:** Differentiate the inner part w.r.t. $w$:
$$\frac{\partial}{\partial w}\left(wx^{(i)} + b - y^{(i)}\right) = x^{(i)}$$

**Step 3:** The 2 from the power rule cancels with the $\frac{1}{2}$:
$$\boxed{\frac{\partial J(w,b)}{\partial w} = \frac{1}{m} \sum_{i=0}^{m-1} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right) x^{(i)}}$$

---

### Deriving $\frac{\partial J}{\partial b}$

**Step 1:** Same chain rule:
$$\frac{\partial J}{\partial b} = \frac{1}{2m} \sum_{i=0}^{m-1} 2\left(wx^{(i)} + b - y^{(i)}\right) \cdot \frac{\partial}{\partial b}\left(wx^{(i)} + b - y^{(i)}\right)$$

**Step 2:** Differentiate the inner part w.r.t. $b$:
$$\frac{\partial}{\partial b}\left(wx^{(i)} + b - y^{(i)}\right) = 1$$

**Step 3:** The 2 cancels the $\frac{1}{2}$:
$$\boxed{\frac{\partial J(w,b)}{\partial b} = \frac{1}{m} \sum_{i=0}^{m-1} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)}$$

---

## Step-by-Step Worked Example (By Hand)

Our training data:

| $x^{(i)}$ | $y^{(i)}$ |
|-----------|-----------|
| 1.0       | 300       |
| 2.0       | 500       |

Let's start with $w = 0$, $b = 0$, $\alpha = 0.1$, and do **one** gradient descent step.

**Step 1: Compute predictions**
$$f(1.0) = 0 \times 1.0 + 0 = 0$$
$$f(2.0) = 0 \times 2.0 + 0 = 0$$

**Step 2: Compute errors** $(f - y)$
$$\text{error}_0 = 0 - 300 = -300$$
$$\text{error}_1 = 0 - 500 = -500$$

**Step 3: Compute $\frac{\partial J}{\partial w}$** (multiply each error by its $x^{(i)}$, then average)
$$\frac{\partial J}{\partial w} = \frac{1}{2}\left[(-300)(1.0) + (-500)(2.0)\right] = \frac{-300 - 1000}{2} = -650$$

**Step 4: Compute $\frac{\partial J}{\partial b}$** (just average the errors)
$$\frac{\partial J}{\partial b} = \frac{1}{2}\left[(-300) + (-500)\right] = \frac{-800}{2} = -400$$

**Step 5: Update $w$ and $b$ simultaneously**
$$w_{\text{new}} = 0 - 0.1 \times (-650) = 0 + 65 = 65$$
$$b_{\text{new}} = 0 - 0.1 \times (-400) = 0 + 40 = 40$$

After just **one step**, $w$ went from 0 to 65 and $b$ went from 0 to 40 — already moving toward the optimal $w=200, b=100$.

We repeat this process thousands of times, each time getting closer to the minimum.

````python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

print("Libraries loaded successfully!")
````

## Part 1: The Cost Function (from the previous notebook)

We need this to track progress during gradient descent — each iteration we check whether cost is going down.

````python
def compute_cost(x, y, w, b):
    """
    Computes J(w, b) — the mean squared error cost.

    Args:
        x : numpy array (m,)  — input features
        y : numpy array (m,)  — true target values
        w : float             — slope (weight)
        b : float             — intercept (bias)

    Returns:
        cost : float
    """
    m = len(x)
    total_squared_error = 0

    for i in range(m):
        prediction = w * x[i] + b
        error = prediction - y[i]
        total_squared_error += error ** 2

    return total_squared_error / (2 * m)


# Quick verification
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"Cost at w=200, b=100 : {compute_cost(x_train, y_train, 200, 100)}")  # Expected: 0.0
print(f"Cost at w=0,   b=0   : {compute_cost(x_train, y_train, 0,   0)}")    # Expected: large
````

---
## Part 2: Implement `compute_gradient`

This translates the two gradient formulas we derived directly into code:

$$\frac{\partial J}{\partial w} = \frac{1}{m} \sum_{i=0}^{m-1} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right) x^{(i)}$$

$$\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=0}^{m-1} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)$$

Each line of code maps directly to one symbol in the formula:

| Math | Code |
|------|------|
| $f_{w,b}(x^{(i)})$ | `f_wb = w * x[i] + b` |
| $f_{w,b}(x^{(i)}) - y^{(i)}$ | `error = f_wb - y[i]` |
| $\text{error} \cdot x^{(i)}$ | `dj_dw += error * x[i]` |
| $\text{error} \cdot 1$ | `dj_db += error` |
| divide by $m$ | `dj_dw /= m`, `dj_db /= m` |

````python
def compute_gradient(x, y, w, b):
    """
    Computes the gradient of J(w, b) w.r.t. w and b.

    Args:
        x : numpy array (m,)  — input features
        y : numpy array (m,)  — true target values
        w : float             — current slope
        b : float             — current intercept

    Returns:
        dj_dw : float — partial derivative of J w.r.t. w
        dj_db : float — partial derivative of J w.r.t. b
    """
    m = len(x)
    dj_dw = 0.0
    dj_db = 0.0

    for i in range(m):
        f_wb  = w * x[i] + b          # current prediction
        error = f_wb - y[i]           # how far off we are

        dj_dw += error * x[i]         # accumulate gradient for w
        dj_db += error                # accumulate gradient for b

    dj_dw /= m                        # average over all m examples
    dj_db /= m

    return dj_dw, dj_db


# --- Verify with our worked example: w=0, b=0 ---
dj_dw, dj_db = compute_gradient(x_train, y_train, w=0, b=0)
print(f"At w=0, b=0:")
print(f"  dj/dw = {dj_dw:.1f}  (expected: -650.0)")
print(f"  dj/db = {dj_db:.1f}  (expected: -400.0)")

# --- At the minimum: gradients should be zero (or near zero) ---
dj_dw_min, dj_db_min = compute_gradient(x_train, y_train, w=200, b=100)
print(f"\nAt w=200, b=100 (optimal):")
print(f"  dj/dw = {dj_dw_min:.6f}  (expected: ≈ 0)")
print(f"  dj/db = {dj_db_min:.6f}  (expected: ≈ 0)")
````

---
## Part 3: Implement `gradient_descent`

This is the main loop that repeatedly applies the update rules:

$$w := w - \alpha \cdot \frac{\partial J}{\partial w}$$
$$b := b - \alpha \cdot \frac{\partial J}{\partial b}$$

We run this for `num_iters` iterations and save the cost at each step so we can visualize the learning progress.

> **Key detail:** Both `w` and `b` must be updated using the gradients computed at the **current** `w` and `b`. Never update `w` first and then use the new `w` when computing `dj_db` — that would be wrong.

````python
def gradient_descent(x, y, w_init, b_init, alpha, num_iters):
    """
    Runs gradient descent to learn optimal w and b.

    Args:
        x        : numpy array (m,)  — input features
        y        : numpy array (m,)  — true target values
        w_init   : float             — starting value for w
        b_init   : float             — starting value for b
        alpha    : float             — learning rate
        num_iters: int               — number of iterations to run

    Returns:
        w         : float            — learned slope
        b         : float            — learned intercept
        J_history : list             — cost at each iteration (for plotting)
        wb_history: list             — [w, b] at each iteration (for plotting)
    """
    J_history  = []    # track cost over time
    wb_history = []    # track (w, b) path over time

    w = w_init
    b = b_init

    for i in range(num_iters):
        # Step 1: Compute gradients at the CURRENT w, b
        dj_dw, dj_db = compute_gradient(x, y, w, b)

        # Step 2: Update w and b SIMULTANEOUSLY
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        # Step 3: Record history (guard against list growing too large)
        if i < 100_000:
            J_history.append(compute_cost(x, y, w, b))
            wb_history.append([w, b])

        # Step 4: Print progress at 10 evenly spaced intervals
        if i % max(1, math.ceil(num_iters / 10)) == 0:
            print(f"Iter {i:5d}  |  Cost: {J_history[-1]:10.4f}  |  "
                  f"w: {w:9.4f}  b: {b:9.4f}  |  "
                  f"dj_dw: {dj_dw:9.4f}  dj_db: {dj_db:9.4f}")

    return w, b, J_history, wb_history
````

---
## Part 4: Run Gradient Descent

We start at $w=0, b=0$ and let gradient descent find the optimal values.

- **Learning rate** $\alpha = 0.01$ — a safe starting value
- **Iterations** = 10,000 — enough to converge for this small dataset

Watch how the cost rapidly drops at first, then slowly flattens as we approach the minimum.

````python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

w_init     = 0
b_init     = 0
alpha      = 1.0e-2   # learning rate
iterations = 10_000

w_final, b_final, J_hist, wb_hist = gradient_descent(
    x_train, y_train, w_init, b_init, alpha, iterations
)

print(f"\nOptimal w = {w_final:.4f}  (target: 200)")
print(f"Optimal b = {b_final:.4f}  (target: 100)")
print(f"Final cost = {J_hist[-1]:.6f}")
````

---
## Part 5: (Graph 1) Cost vs. Iterations

### What this graph shows:
- The **x-axis** is the iteration number (how many steps gradient descent has taken)
- The **y-axis** is the cost at that step
- A good learning rate produces a **steadily decreasing** curve
- The curve is shown in **two panels** because the drop is so fast at the start that the later progress is invisible on a single scale

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# Left: first 200 iterations — the rapid initial drop
ax1.plot(J_hist[:200], color='steelblue', linewidth=2)
ax1.set_title('Cost vs. Iterations (first 200)', fontsize=13)
ax1.set_xlabel('Iteration', fontsize=11)
ax1.set_ylabel('Cost J(w, b)', fontsize=11)
ax1.grid(True, alpha=0.3)

# Right: from iteration 1000 onward — the slow final approach
ax2.plot(np.arange(1000, len(J_hist)), J_hist[1000:], color='darkorange', linewidth=2)
ax2.set_title('Cost vs. Iterations (from 1000 onward)', fontsize=13)
ax2.set_xlabel('Iteration', fontsize=11)
ax2.set_ylabel('Cost J(w, b)', fontsize=11)
ax2.grid(True, alpha=0.3)

plt.suptitle('Gradient Descent — Cost History', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"Starting cost : {J_hist[0]:,.2f}")
print(f"Final cost    : {J_hist[-1]:.6f}")
````

## Part 6: (Graph 2) Gradient Descent Path on the Contour Map

### What this graph shows:
- The **contour rings** show the cost $J(w, b)$ for every combination of $w$ and $b$
- Each ring connects points with the **same cost value** — like elevation lines on a map
- **Inner rings** = lower cost = better parameters
- The **red dots** are the actual positions of $(w, b)$ at each gradient descent step
- The dots spiral **inward** toward the center — the minimum
- Notice: steps are **large at the start** and **tiny near the end**, because the gradient shrinks as we approach the minimum

````python
# Build the cost grid for the contour background
w_vals = np.linspace(0,   400, 100)
b_vals = np.linspace(-50, 250, 100)
W, B   = np.meshgrid(w_vals, b_vals)
J_grid = np.array([
    [compute_cost(x_train, y_train, w, b) for w in w_vals]
    for b in b_vals
])

# Extract the (w, b) path taken by gradient descent
ws = [p[0] for p in wb_hist]
bs = [p[1] for p in wb_hist]

# Sample every 500th point so the plot stays readable
step = 500
ws_sampled = ws[::step]
bs_sampled = bs[::step]

fig, ax = plt.subplots(figsize=(9, 7))

# Contour background
cp = ax.contour(W, B, J_grid, levels=40, cmap='viridis')
plt.colorbar(cp, ax=ax, label='Cost J(w, b)')

# Gradient descent path
ax.plot(ws_sampled, bs_sampled, 'ro-', markersize=5, linewidth=1.5,
        label='Gradient descent path')
ax.plot(ws_sampled[0],  bs_sampled[0],  'g^', markersize=12, label=f'Start  (w={ws[0]:.0f}, b={bs[0]:.0f})')
ax.plot(ws_sampled[-1], bs_sampled[-1], 'b*', markersize=14, label=f'End    (w={w_final:.1f}, b={b_final:.1f})')

ax.set_xlabel('w  (slope)', fontsize=12)
ax.set_ylabel('b  (intercept)', fontsize=12)
ax.set_title('Gradient Descent Path on Cost Contour Map', fontsize=14)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
````

## Part 9: The Learning Rate $\alpha$ — Why It Matters

The learning rate $\alpha$ is one of the most important choices you make when running gradient descent.

| $\alpha$ is... | Effect |
|----------------|--------|
| **Too small** | Gradient descent works but takes a very long time to converge |
| **Just right** | Converges smoothly and efficiently |
| **Too large** | Overshoots the minimum — $w$ and $b$ bounce back and forth, cost **increases** |

Let's see all three cases in action.

````python
configs = [
    (1.0e-4, 10_000, 'Too small  (α = 1e-4)',  'steelblue'),
    (1.0e-2, 10_000, 'Just right (α = 1e-2)',  'green'),
    (9.0e-1, 10,     'Too large  (α = 9e-1)',   'crimson'),
]

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ax, (alpha_val, iters, label, color) in zip(axes, configs):
    _, _, j_hist, _ = gradient_descent(x_train, y_train, 0, 0, alpha_val, iters)
    ax.plot(j_hist, color=color, linewidth=2)
    ax.set_title(label, fontsize=12)
    ax.set_xlabel('Iteration', fontsize=10)
    ax.set_ylabel('Cost J(w, b)', fontsize=10)
    ax.grid(True, alpha=0.3)

plt.suptitle('Effect of Learning Rate on Gradient Descent', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

### Reading the graphs:
- **Blue (too small):** Cost decreases but very slowly — you would need far more iterations
- **Green (just right):** Rapid smooth descent — this is what we want
- **Red (too large):** Cost **goes up** — gradient descent is diverging, overshooting the minimum every step

A good rule of thumb for choosing $\alpha$: start at `0.01` and adjust. If cost grows → lower $\alpha$. If convergence is too slow → slightly raise $\alpha$.

## Part 10: Step-by-Step Trace of One Run

Let's print every computed value for the **first 5 iterations** to make gradient descent completely transparent.

````python
w, b = 0.0, 0.0
alpha = 0.01

print(f"{'Iter':>5} {'w':>10} {'b':>10} {'dj/dw':>12} {'dj/db':>12} {'Cost':>14}")
print("-" * 65)

for i in range(6):
    dj_dw, dj_db = compute_gradient(x_train, y_train, w, b)
    cost = compute_cost(x_train, y_train, w, b)
    print(f"{i:>5} {w:>10.4f} {b:>10.4f} {dj_dw:>12.4f} {dj_db:>12.4f} {cost:>14.4f}")
    # Simultaneous update
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
````

## Part 11: Apply to a Different Dataset

Now let's use gradient descent on a **different real-world-style dataset**: student study hours vs. exam scores.

The exact same `compute_gradient` and `gradient_descent` functions work on **any** linear regression problem.

| Study Hours ($x$) | Exam Score ($y$) |
|------------------|------------------|
| 1                | 52               |
| 2                | 58               |
| 3                | 65               |
| 4                | 71               |
| 5                | 75               |
| 6                | 82               |
| 7                | 88               |

````python
x_study = np.array([1, 2, 3, 4, 5, 6, 7], dtype=float)
y_score = np.array([52, 58, 65, 71, 75, 82, 88], dtype=float)

w_s, b_s, J_s, wb_s = gradient_descent(
    x_study, y_score,
    w_init=0, b_init=0,
    alpha=0.01,
    num_iters=10_000
)

print(f"\nLearned: w = {w_s:.4f}, b = {b_s:.4f}")
print(f"Final cost = {J_s[-1]:.4f}")
````

````python
# Visualize the result
x_line_s = np.linspace(0.5, 7.5, 100)
y_line_s  = w_s * x_line_s + b_s

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: prediction vs data
ax1.plot(x_line_s, y_line_s, color='steelblue', linewidth=2,
         label=f'f(x) = {w_s:.2f}x + {b_s:.2f}')
ax1.scatter(x_study, y_score, color='black', s=80, zorder=5, label='Actual scores')
for xi, yi in zip(x_study, y_score):
    ax1.plot([xi, xi], [yi, w_s * xi + b_s], 'r--', linewidth=1)
ax1.plot([], [], 'r--', label='Residuals')
ax1.set_title(f'Study Hours vs Exam Score\nCost = {J_s[-1]:.4f}', fontsize=12)
ax1.set_xlabel('Study Hours', fontsize=11)
ax1.set_ylabel('Exam Score', fontsize=11)
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3)

# Right: cost history
ax2.plot(J_s, color='darkorange', linewidth=2)
ax2.set_title('Cost vs. Iterations', fontsize=12)
ax2.set_xlabel('Iteration', fontsize=11)
ax2.set_ylabel('Cost J(w, b)', fontsize=11)
ax2.grid(True, alpha=0.3)

plt.suptitle('Gradient Descent on Study Hours Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

````python
# Step-by-step cost breakdown at the learned parameters
m = len(x_study)
print(f"{'i':<5} {'x':>8} {'y actual':>10} {'f(x) pred':>12} {'error':>10} {'error²':>12}")
print("-" * 60)

total_sq = 0
for i in range(m):
    pred  = w_s * x_study[i] + b_s
    error = pred - y_score[i]
    sq    = error ** 2
    total_sq += sq
    print(f"{i:<5} {x_study[i]:>8.1f} {y_score[i]:>10.1f} {pred:>12.2f} {error:>10.2f} {sq:>12.2f}")

print("-" * 60)
print(f"{'Sum of squared errors':>45} {total_sq:>12.2f}")
print(f"{'Cost  J = sum / (2m) =':>45} {total_sq / (2 * m):>12.4f}")

print(f"\nPredictions for new inputs:")
print(f"  3.5 study hours → score ≈ {w_s * 3.5 + b_s:.1f}")
print(f"  8.0 study hours → score ≈ {w_s * 8.0 + b_s:.1f}")
````

## Summary: Key Takeaways

| Concept | In Plain English |
|---------|------------------|
| **Gradient** | The slope of the cost function — tells you which direction is "uphill" |
| **Gradient Descent** | Repeatedly step downhill until you reach the minimum |
| **Learning rate $\alpha$** | Controls the size of each step |
| **Too small $\alpha$** | Correct direction, very slow convergence |
| **Too large $\alpha$** | Overshoots the minimum — cost diverges |
| **$\frac{\partial J}{\partial w}$** | How much the cost changes when $w$ changes slightly |
| **$\frac{\partial J}{\partial b}$** | How much the cost changes when $b$ changes slightly |
| **Simultaneous update** | Compute both gradients before updating either parameter |
| **Convergence** | When gradients ≈ 0 and cost stops decreasing |
| **Cost history plot** | Monotonically decreasing = correct; increasing = $\alpha$ too large |
| **Contour path** | Spirals inward toward the minimum — large steps at start, tiny near the end |
